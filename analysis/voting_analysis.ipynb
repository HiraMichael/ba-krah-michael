# Descriptive statistics

```python
def compute_descriptive_statistics(df, column_name):
    stats = {
        'mean': df[column_name].mean(),
        'std': df[column_name].std(),
        '25%': df[column_name].quantile(0.25),
        '50% (median)': df[column_name].quantile(0.50),
        '75%': df[column_name].quantile(0.75),
        'min': df[column_name].min(),
        'max': df[column_name].max()
    }
    return stats
```

```python
labels = ['Equality', 'Effort', 'Needs', 'Productivity', 'Social Utility', 'Supply and Demand']
vr1_dfs = []
for i, (canon, group) in enumerate(vr.groupby('canon')):
    desc_stats = compute_descriptive_statistics(group, 'weights')
    vr1_dfs.append(pd.DataFrame(desc_stats, index=[f'NS {labels[i]}']))
vr1_desc_stats = pd.concat([df for df in vr1_dfs])
display(vr1_desc_stats)
```

```python
vr2_dfs = []
for i, (canon, group) in enumerate(vr2.groupby('canon')):
    desc_stats = compute_descriptive_statistics(group, 'weights')
    vr2_dfs.append(pd.DataFrame(desc_stats, index=[f'MM {labels[i]}']))
vr2_desc_stats = pd.concat([df for df in vr2_dfs])
display(vr2_desc_stats)
```

```python
voting_desc_stats = pd.concat([vr1_desc_stats, vr2_desc_stats]).round(3)
voting_desc_stats = voting_desc_stats.applymap(lambda x: f"{x:.3f}".rstrip('0').rstrip('.') if isinstance(x, (int, float)) else x)
#display(voting_desc_stats)
voting_desc_stats.to_latex(path + '/analysis/tables/voting_desc_stats.tex')
```

### Narrow self-interest

# Agent-specific statistics

```python
car = pd.read_csv(path + '/results/EXP1_canon_allocation_results.csv', index_col=[0,1])
```

```python
agent_cols = [f'agent{i}' for i in range(111)]
time_steps = sorted(vr.index.get_level_values('time').unique())
agent_distances_df = pd.DataFrame(index=agent_cols, columns=time_steps, dtype=np.float32)
max_sn_mva = 0.63
max_loading = 0.85
for time, group in vr.groupby('time'):
    for agent in agent_cols:
        agent_weights = group[agent].values
        
        car_at_time = car.loc[time]
        sod_columns = [
            'sod_effort', 'sod_equality', 'sod_needs',
            'sod_social_utility', 'sod_productivity', 'sod_supply_and_demand'
        ]
        
        agent_preferred_sod = np.zeros(len(car_at_time))
        for i, canon_sod in enumerate(sod_columns):
            agent_preferred_sod += agent_weights[i] * car_at_time[canon_sod]
        
        # adjust
        alpha = (max_sn_mva * max_loading) / np.sum(agent_preferred_sod * car_at_time['demand'])
        agent_preferred_sod = np.minimum(alpha * agent_preferred_sod, 1)
        
        distance_to_actual_sod = compute_euclidean_distance(agent_preferred_sod, car_at_time['share_of_demand'])
        agent_distances_df.at[agent, time] = np.float32(distance_to_actual_sod)
        
display(agent_distances_df.head(20))
```

```python
agent_distances_df.to_csv(path + '/analysis/exp1_agent_distances.csv')
```

```python
distance_columns = [f'agent{i}' for i in range(111)]

euclidean_distance_df = pd.DataFrame(columns=distance_columns)

for time in vr.index.get_level_values('time').unique():
    vr_at_time = vr.loc[time]
    
    votes_columns = [f'agent{i}' for i in range(111)]
    
    vectors = vr_at_time[votes_columns].values
    
    canon_weights = vr_at_time['weights'].values
    
    distances = [
        compute_euclidean_distance(vectors[:, i], canon_weights)
        for i in range(vectors.shape[1])
    ]
    
    # Append the results for the current time step to the result dataframe
    euclidean_distance_df.loc[time] = distances
```
